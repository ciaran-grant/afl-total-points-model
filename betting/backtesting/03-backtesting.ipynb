```python
backtest_data['win_Open'] = np.where((backtest_data['Market'] == "Under") & (backtest_data['Total_Game_Score'] < backtest_data['Total_Score_Open']), 1, 
                            np.where((backtest_data['Market'] == "Over") & (backtest_data['Total_Game_Score'] > backtest_data['Total_Score_Open']), 1,
                                     0))
backtest_data['win_Close'] = np.where((backtest_data['Market'] == "Under") & (backtest_data['Total_Game_Score'] < backtest_data['Total_Score_Close']), 1, 
                            np.where((backtest_data['Market'] == "Over") & (backtest_data['Total_Game_Score'] > backtest_data['Total_Score_Close']), 1,
                                     0))
```

Betting Functions for Backtesting

Betting when predicted odds are lower (higher probability) than true betting odds.

```python
backtest_data['bet'] = np.where(backtest_data['Total_Score_Close_Pred_Odds'] < backtest_data['Total_Score_Close_True_Odds'], 1, 0)

stake = 1
backtest_data['stake'] = np.where(backtest_data['bet'] == 0, 0, stake)

backtest_data['gpl'] = np.where(backtest_data['bet'] == 1, 
                                np.where(backtest_data['win_Close'] == 1, 
                                         backtest_data['stake'] * (backtest_data['Total_Score_Close_Odds']-1), 
                                         -backtest_data['stake']), 
                                0
                    ) 

backtest_data.head()
```

Calculating a commission

```python
backtest_data.shape, len(backtest_data['Match_ID'].unique())
```

```python
# Apply 5% commission
backtest_data['commission'] = np.where(backtest_data['gpl'] <= 0,
                                       0, 
                                       0.05 * backtest_data['gpl'])

backtest_data['npl'] = backtest_data['gpl'] - backtest_data['commission']

backtest_data.tail()
```

Betting Evaluation Functions

```python
pd.DataFrame(backtest_data[['npl', 'stake']].sum()).transpose().assign( pot = lambda x: x['npl'] / x['stake'])
```

```python
# Cumulative PL by market to visually see trend and consistency
def bet_eval_chart_cPl(d):

    d = (
        d
        .groupby('market_id')
        .agg({'npl': 'sum'})
    )

    d['market_number'] = np.arange(len(d))
    d['cNpl'] = d.npl.cumsum()

    chart = px.line(d, x="market_number", y="cNpl", title='Cumulative Net Profit', template='simple_white')

    return(chart)
```

```python
match_npl = backtest_data.groupby('Match_ID').agg({'npl': 'sum'})
match_npl['match_number'] = np.arange(len(match_npl)) +1
match_npl['cumulative_npl'] = match_npl['npl'].cumsum()
# match_npl
chart = px.line(match_npl, x="match_number", y="cumulative_npl", title='Cumulative Net Profit', template='simple_white')
chart
```
